```python
#Regresion Ridge
alpha = 10**np.linspace(-10, 2, 500)
error_array = np.zeros(len(alpha))
for idx, a in enumerate(alpha):
  ridge_regression = Ridge(alpha=a)
  ridge_regression.fit(x_train, y_train)
  y_predRidge = ridge_regression.predict(x_test)
  error_array[idx] = mean_squared_error(y_test, y_predRidge)
```

```python
# El mean squared error!!!!
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_predRidge))

# El mean absolute error!!!!
print('Mean absolute error: %.2f'
      % mean_absolute_error(y_test, y_predRidge))

# El R2!!!!
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_predRidge))
```

```python
smallest_error = np.min(error_array)
optimal_alpha = alpha[np.argmin(error_array)]
print(f'El valor óptimo de alpha es: {np.round(optimal_alpha, 2)}, con el que se obtiene un error de {np.round(smallest_error, 2)}')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10, 7))
x = np.array(df_clus)
plt.scatter(x[:,0], x[:,1])
plt.xlabel("Monto facturado en compras nacionales en el mes 12")
plt.ylabel("Numero de transacciones en compras nacionales en el mes 12")
plt.show()
```

```python
# Importo el método de clustering jerárquico (bottom-up)
from sklearn.cluster import AgglomerativeClustering
# importamos el puntaje de silhouette
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients = []

# Se necesita tener al menos 2 clusters y a los sumo N-1 (con N el numero de muestras) para obtener coeficientes de Silohuette
for k in range(2, 20):
     cluster = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
     cluster.fit(df_clus)
     score = silhouette_score(df_clus, cluster.labels_)
     silhouette_coefficients.append(score)
```

```python
fig, ax = plt.subplots(figsize = (24, 7))

# estas lineas son el grafico de SSE vs K
ax.scatter(range(2, 20), silhouette_coefficients)            
ax.set_xticks(range(2, 20))
ax.set_xlabel("Número de clusters")
ax.set_ylabel("Promedio coeficientes de Silhouette")
```

```python
# Creamos el modelo
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')

# Lo ajustamos con los datos
cluster.fit_predict(df_clus) 
```

```python
colores = ['red','purple']
asignar=[]
for pre in cluster.labels_:
  asignar.append(colores[pre])
plt.figure(figsize=(10, 7))
x = np.array(df_clus)
plt.scatter(x[:,0], x[:,1], c=asignar)
plt.xlabel("Monto facturado en compras nacionales en el mes 12")
plt.ylabel("Numero de transacciones en compras nacionales en el mes 12")
plt.show()
```

Al aplicar Clustering Jerarquico, se puede apreciar efectivamente los dos grupos de clientes por su comportamiento del Banco Monopoly, por lo que, podemos interpretar que el grupo morado son los clientes propensos a dejar el banco, puesto que tienen pocas transacciones y su monto facturado es bajo, mientras que el grupo rojo posee altas transacciones y un alto monto de facturación, siendo los clientes más activos del banco.
